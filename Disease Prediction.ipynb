```python
import os 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings
import pickle
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.naive_bayes import GaussianNB 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
from xgboost import XGBClassifier
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_csv('data/Training.csv')
```

```python
train_data.isnull().sum()
```

```python
train_data.drop('Unnamed: 133',axis = 1,inplace =True)
```

```python
train_data.head(5)
```

```python
train_data['prognosis'].value_counts()
```

```python
unique_values = train_data['prognosis'].unique()
```

```python
unique_values
```

```python
values_dict = {value:idx for idx,value in enumerate(list(unique_values))}
values_dict
```

```python
train_data ['prognosis'] = train_data['prognosis'].map(values_dict)
```

```python
x_data = train_data.drop('prognosis',axis = 1)
y_data = train_data['prognosis']
```

```python
x_data.head(5)
```

```python
y_data.head(5)
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_data.values,y_data.values,train_size=0.8,random_state=21)
```

```python
def train_model(model_name, model, x_train, y_train, x_test, y_test):
    classifier = model
    classifier.fit(x_train, y_train)
    score = classifier.score(x_test, y_test) 
    print(f"The accuracy score of {model_name} is {score}")
    predictions = classifier.predict(x_test) 
    cf_report = classification_report(y_test, predictions)
    print(f"The classification report of {model_name} is:\n", cf_report)
    with open(f"{model_name}_model.pkl", "wb") as model_file:
        pickle.dump(classifier, model_file)
        print(f"{model_name} saved as {model_name}_model.pkl")
    
    return predictions
models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "NaiveBayes": GaussianNB(),
    "KNN": KNeighborsClassifier()
}
all_model_predictions = {}
for model_name, model in models.items():
    predictions = train_model(model_name, model, X_train, y_train, X_test, y_test)
    all_model_predictions[model_name] = predictions
```
